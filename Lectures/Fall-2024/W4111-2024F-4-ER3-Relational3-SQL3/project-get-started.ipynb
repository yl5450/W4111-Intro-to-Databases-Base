Character is a reserved word in SQL. So, I am going to change it.

```python
%%sql

use lor_data;

alter table characters_raw
    change `character` character_name text null;

alter table interactions_raw
    change `character` character_name text null;
```

```python
%sql select * from characters_raw;
```

```python
%%sql

with one as (select
        book as main_book,
        chapter_location,
    substr(chapter_location, 1, locate(',', chapter_location)-1) as sub_book,
    substr(chapter_location, locate(',', chapter_location)+1) as the_rest
from
    characters_raw),
two as (select
    main_book,
    if(sub_book = '', 'Book 1', sub_book) as sub_book,
    substr(the_rest, 1, locate(':', the_rest)-1) chapter_no,
    substr(the_rest, locate(':', the_rest)+1) chapter_title
from
    one)
select
    main_book,
    substr(sub_book, 6) as sub_book_no,
    substr(trim(chapter_no), length('Chapter ')+1) as chapter_no,
    chapter_title
from
    two;
```
